```python
#At this point yes, I did this using a bit of help, 
#cause afcourse how lengthy this is after doing 4,5,6 :'(

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Part I: Randomized Sales Data Generation
roll_number = 12345  # Change this to your roll number
np.random.seed(roll_number)

# Generate sales data for 12 months (4 categories)
sales_data = np.random.randint(1000, 5000, size=(12, 4))
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(sales_data, columns=categories, index=months)

# Part II: Data Manipulation and Analysis
print(df.head())  # First 5 rows
print(df.describe())  # Summary statistics

df["Total Sales"] = df.sum(axis=1)  # Total sales per month
df["Growth Rate"] = df["Total Sales"].pct_change() * 100  # Percentage change

even_roll = roll_number % 2 == 0
if even_roll:
    df["Electronics"] *= 0.9  # 10% discount
else:
    df["Clothing"] *= 0.85  # 15% discount

# Part III: Visualizations
plt.figure(figsize=(12, 5))
sns.lineplot(data=df[categories])
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(categories)
plt.grid()
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.grid()
plt.show()

# Q2: Array Operations
array = np.array([[1, -2, 3], [-4, 5, -6]])
print("Absolute Value:", np.abs(array))

print("Percentiles:")
print("Flattened 25%:", np.percentile(array, 25))
print("Flattened 50%:", np.percentile(array, 50))
print("Flattened 75%:", np.percentile(array, 75))

print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std Dev:", np.std(array))

# Q3: Floor, Ceiling, Truncated Values
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Truncated:", np.trunc(a))
print("Rounded:", np.round(a))

# Q4: Swap elements using a temporary variable
lst = [1, 2, 3, 4, 5]
i, j = 1, 3
temp = lst[i]
lst[i] = lst[j]
lst[j] = temp
print("Swapped List:", lst)

# Q5: Swap elements in a set
s = {10, 20, 30, 40}
s_lst = list(s)
s_lst[0], s_lst[1] = s_lst[1], s_lst[0]
s = set(s_lst)
print("Swapped Set:", s)
```
