```python
# Importing all the serious libraries, but my attendance is still a joke
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from wordcloud import WordCloud
from textblob import TextBlob
import matplotlib.pyplot as plt

# Downloading NLTK data, because unlike my attendance, these can't be skipped
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Q1: The paragraph about my favorite topic (attendance hacks)
attendance_hack_paragraph = """
In college, attending classes is an art. Sometimes, the real skill is not in learning, but in collecting attendance links from friends. 
The teacher, who rules both the lab and lecture, occasionally grants magical free attendance. 
This creates a secret society of students, united by the sacred sharing of attendance links. 
While some attend physically, others attend spiritually (via WhatsApp). 
In the end, everyone passes, but only a few master the art of proxy presence.
"""

# 1. Convert to lowercase and remove punctuation (because my attendance is already low, let's lower the case too)
attendance_lowered = attendance_hack_paragraph.lower()
attendance_no_punct = re.sub(r'[^\w\s]', '', attendance_lowered)  # Punctuation removed, like my worries

# 2. Tokenize into words and sentences (splitting my attendance between lab and lecture)
attendance_words = word_tokenize(attendance_no_punct)
attendance_sentences = sent_tokenize(attendance_hack_paragraph)

# 3. Split using split() vs word_tokenize()
attendance_split_words = attendance_no_punct.split()
attendance_word_tokenize = word_tokenize(attendance_no_punct)
# Comparing: split() is like just forwarding the link, word_tokenize() is like actually attending

# 4. Remove stopwords (because "the", "is", and "in" don't help my attendance)
stop_words = set(stopwords.words('english'))
attendance_no_stopwords = [word for word in attendance_word_tokenize if word not in stop_words]

# 5. Word frequency distribution (excluding stopwords, because who wants extra noise in attendance)
from collections import Counter
attendance_freq_dist = Counter(attendance_no_stopwords)

print("Q1 Results:")
print("Lowercased + No Punctuation:", attendance_no_punct)
print("Words (split):", attendance_split_words)
print("Words (word_tokenize):", attendance_word_tokenize)
print("Difference: split() doesn't handle punctuation, word_tokenize() does (like a teacher who actually checks the list)")
print("Words w/o stopwords:", attendance_no_stopwords)
print("Word Frequency:", attendance_freq_dist)
print()

# Q2: More attendance filtering
# 1. Extract all words with only alphabets (no roll numbers, just names)
attendance_alpha_words = re.findall(r'\b[a-zA-Z]+\b', attendance_no_punct)

# 2. Remove stopwords again (double proxy)
attendance_alpha_no_stop = [word for word in attendance_alpha_words if word not in stop_words]

# 3. Stemming (cutting words like cutting attendance)
stemmer = PorterStemmer()
attendance_stemmed = [stemmer.stem(word) for word in attendance_alpha_no_stop]

# 4. Lemmatization (making words proper, like making my attendance proper)
lemmatizer = WordNetLemmatizer()
attendance_lemmatized = [lemmatizer.lemmatize(word) for word in attendance_alpha_no_stop]

# 5. Compare stemmed vs lemmatized
print("Q2 Results:")
print("Alphabetic Words:", attendance_alpha_words)
print("No Stopwords:", attendance_alpha_no_stop)
print("Stemmed:", attendance_stemmed)
print("Lemmatized:", attendance_lemmatized)
print("When to use stemming: When you want quick and dirty (like proxy attendance).")
print("When to use lemmatization: When you want correctness (like actually attending).")
print()

# Q3: Bag of Words and TF-IDF (because my attendance is everywhere)
texts_for_vectorizer = [
    "Teacher gives free attendance in lab.",
    "Everyone shares the attendance link with me.",
    "Proxy attendance is the real college skill."
]

# CountVectorizer
vectorizer_attendance = CountVectorizer()
attendance_bow = vectorizer_attendance.fit_transform(texts_for_vectorizer)
print("Q3 Bag of Words feature names:", vectorizer_attendance.get_feature_names_out())
print("Bag of Words Matrix:\n", attendance_bow.toarray())

# TfidfVectorizer
tfidf_vectorizer_attendance = TfidfVectorizer()
attendance_tfidf = tfidf_vectorizer_attendance.fit_transform(texts_for_vectorizer)
print("TF-IDF feature names:", tfidf_vectorizer_attendance.get_feature_names_out())
print("TF-IDF Matrix:\n", attendance_tfidf.toarray())

# Top 3 keywords for each text using TF-IDF (like top 3 students who always get attendance)
for i, text in enumerate(texts_for_vectorizer):
    tfidf_scores = attendance_tfidf[i].toarray()[0]
    top3_idx = tfidf_scores.argsort()[-3:][::-1]
    top3_words = [tfidf_vectorizer_attendance.get_feature_names_out()[j] for j in top3_idx]
    print(f"Text {i+1} Top 3 Keywords:", top3_words)
print()

# Q4: Comparing two technologies (like comparing lab vs lecture attendance)
tech_attendance_lab = "Lab attendance is often more relaxed. Teachers sometimes give free attendance to everyone. The environment is friendly and practical."
tech_attendance_lecture = "Lecture attendance is strictly monitored. Teachers rarely grant free attendance. The environment is formal and theoretical."

# Preprocess and tokenize
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return [word for word in word_tokenize(text) if word not in stop_words]

lab_tokens = set(preprocess(tech_attendance_lab))
lecture_tokens = set(preprocess(tech_attendance_lecture))

# Jaccard Similarity
jaccard_sim = len(lab_tokens & lecture_tokens) / len(lab_tokens | lecture_tokens)

# Cosine Similarity
tfidf_vec = TfidfVectorizer()
tfidf_matrix = tfidf_vec.fit_transform([tech_attendance_lab, tech_attendance_lecture])
cosine_sim = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

print("Q4 Results:")
print("Lab tokens:", lab_tokens)
print("Lecture tokens:", lecture_tokens)
print("Jaccard Similarity:", jaccard_sim)
print("Cosine Similarity:", cosine_sim)
print("Jaccard is like checking who actually attended both lab and lecture. Cosine is like checking who just copied the attendance list. Cosine gives more nuanced similarity here.")
print()

# Q5: Sentiment Analysis (because my attendance is always positive)
review_attendance = "Getting free attendance from my teacher made my day! The process was smooth and everyone was happy."

blob = TextBlob(review_attendance)
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

# Classify review
if polarity > 0.1:
    sentiment = "Positive"
elif polarity < -0.1:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

print("Q5 Results:")
print("Review:", review_attendance)
print("Polarity:", polarity)
print("Subjectivity:", subjectivity)
print("Sentiment:", sentiment)

# Word cloud for positive reviews (because my attendance is floating in the cloud)
if sentiment == "Positive":
    wordcloud = WordCloud(width=400, height=200, background_color='white').generate(review_attendance)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("Word Cloud of Positive Attendance Reviews")
    plt.show()

# Q6: Text Generation (like generating excuses for not attending)
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
import numpy as np

# My own paragraph (about attendance, of course)
attendance_story = """In college, getting attendance is an art form. Some students attend every class, while others master the skill of proxy attendance. Teachers sometimes grant free attendance, making everyone happy. The secret society of link sharers grows stronger each semester. In the end, everyone passes, but only a few become legends."""

# Tokenize
tokenizer_attendance = Tokenizer()
tokenizer_attendance.fit_on_texts([attendance_story])
total_words = len(tokenizer_attendance.word_index) + 1

# Create input sequences
input_sequences = []
for line in attendance_story.split('.'):
    token_list = tokenizer_attendance.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences
max_seq_len = max([len(seq) for seq in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')

# Split into X and y
X_attendance = input_sequences[:, :-1]
y_attendance = input_sequences[:, -1]

# One-hot encode y
from keras.utils import to_categorical
y_attendance = to_categorical(y_attendance, num_classes=total_words)

# Build simple LSTM model (because my attendance is recurrently missing)
model_attendance = Sequential()
model_attendance.add(Embedding(total_words, 10, input_length=max_seq_len-1))
model_attendance.add(LSTM(50))
model_attendance.add(Dense(total_words, activation='softmax'))
model_attendance.compile(loss='categorical_crossentropy', optimizer='adam')

# Train (just a few epochs, like attending only a few classes)
model_attendance.fit(X_attendance, y_attendance, epochs=100, verbose=0)

# Generate text (like generating attendance excuses)
def generate_attendance_text(seed_text, next_words=3):
    for _ in range(next_words):
        token_list = tokenizer_attendance.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')
        predicted = np.argmax(model_attendance.predict(token_list, verbose=0), axis=-1)
        output_word = ""
        for word, index in tokenizer_attendance.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text

print("Q6 Results:")
print("Generated text:", generate_attendance_text("attendance"))
```
